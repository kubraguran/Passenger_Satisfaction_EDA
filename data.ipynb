#step 3 - find the missing values

```python
#how many of data is missing
check_missing_data = df.isnull().sum()

#missing data's percantage 
check_missing_data_percantage = (check_missing_data / len(df)) * 100

#print missing datas
result_of_missings = pd.DataFrame({"Missing Value ": check_missing_data, "Percentage ": check_missing_data_percantage})
print(result_of_missings)
```

```python
#number of missing values in "Arrival delay in mintues" and try to find out group of it
categoricals = df.select_dtypes(include='object').columns
none_values = df[df['Arrival Delay in Minutes'].isnull()]

##check all the groups for "Arrival delay in minutes"

for col in categoricals:
    print(f"According to {col} row\n")
    print(none_values[col].value_counts())
```

```python
#percentage of missing values in "Arrival delay in mintues" and try to find out group of it

for col in categoricals:
    ratio = df[df['Arrival Delay in Minutes'].isnull()][col].value_counts() / df[col].value_counts()
    print(f"missing values for {col}")
    print(ratio.sort_values(ascending=False))
```

According to the missing value analysis, we have missing values only in the "Arrival Delay in Minutes" column.
There are 310 missing entries, which accounts for approximately 0.30% of the entire dataset.

Additionally, we examined the distribution of missing values across categorical variables such as Satisfaction, Class, Type of Travel, etc.
This grouping-based breakdown helps identify whether the missingness is random or systematically related to specific passenger groups.

What we can do for missing values

1- Deletion of Missing Values:
We can delete the missing values if the number of missing value is not too high or if this delete will not effect the our analysis.

2- Filling with Mean or Median:
We can fill out with the mean(average) or median(median).

3- Group-Based Imputation:
 We can fill out based on the group section. First we need to find median for each group and then we can fill out the sections with them

4- Model-Based Imputation:
We can choose model-based imputation. Missing values can also be estimated using advanced regression or other predictive models.

```python
#Model-Based imputation
from sklearn.ensemble import RandomForestRegressor


null_index = df[df['Arrival Delay in Minutes'].isnull()].index

if len(null_index) == 0:
    print("empty rows")
else:
#First find missing and non-missing values
    df_notnull = df[df['Arrival Delay in Minutes'].notnull()]
    df_null = df[df['Arrival Delay in Minutes'].isnull()]
 

#make it numeric to categorial rows
    df_encoded = pd.get_dummies(df, columns=categoricals, drop_first=True)

#make it separate for missing and nonmissing
    df_notnull_encode = df_encoded.loc[df_notnull.index]
    df_null_encode = df_encoded.loc[null_index]


#find targets
    X_train = df_notnull_encode.drop(columns=['Arrival Delay in Minutes'])
    Y_train = df_notnull_encode['Arrival Delay in Minutes']
    X_pred = df_null_encode.drop(columns=['Arrival Delay in Minutes'])

#Creata modele and train it

    model = RandomForestRegressor(n_estimators=100,random_state=45)
    model.fit(X_train,Y_train)

#Predict the missing values
    predicted_value = model.predict(X_pred)

#With this True or False we can train our data better because data will know if original value was null or not
    df['Arrival Delay Imputed'] = False

#Replace it with null and True
    df.loc[null_index, 'Arrival Delay in Minutes'] = predicted_value
    df.loc[null_index, 'Arrival Delay Imputed'] = True
    print(df['Arrival Delay in Minutes'])

```

```python
#check what we did
df[df['Arrival Delay Imputed'] == True].head()
```

```python
#Check the before and after for missing values
print("Before Imputation")
print(df.loc[null_index, 'Arrival Delay in Minutes'].describe())

print("After Imputation")
print(df['Arrival Delay in Minutes'].describe())
```

```python
#create scatter plots
plt.figure(figsize=(8,6))
sns.histplot(predicted_value, bins=30, kde=True, color='orange', label='Predicted Values')
plt.title('Distribution of Predicted (Imputed) Values')
plt.legend()
plt.show()
 
```

```python
# Outlier Analysis

Q1 = df['Arrival Delay in Minutes'].quantile(0.25)
Q3 = df['Arrival Delay in Minutes'].quantile(0.75)

IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

print(f'Upper bound for outlier is {upper_bound}')
print(f"Lower bound for outlier is {lower_bound}")

outliers = df[(df['Arrival Delay in Minutes'] < lower_bound) | (df['Arrival Delay in Minutes'] > upper_bound)]
print(f"Total number of outlier is {len(outliers)}")
```

```python
#what is the percentage of outlier values
outlier_ratio = len(outliers) / len(df) * 100
print(f"Percentage of outlier is {round(outlier_ratio), 2}")
```
